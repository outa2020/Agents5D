# 4. Methodology & Implementation

## ğŸ“Š Data Architecture

We use **strongly-typed data models** for reliable agent communication. This implements **ADK Concept #5: Structured Outputs**.

### ğŸ¯ Why Structured Data Matters

**Without structured outputs:**
- âŒ Agents return unpredictable text
- âŒ Hard to parse and validate
- âŒ Errors cascade through system

**With structured outputs (our approach):**
- âœ… Type-safe dataclasses
- âœ… Predictable JSON format

- âœ… Easy agent-to-agent communication

---

### ğŸ“¥ Input Model: DisasterReport

**Purpose:** Standardize incoming disaster information

**Fields:**
- `report_id`: Unique identifier (e.g., "DS-2024-001")
- `timestamp`: When disaster occurred
- `location`: Geographic location
- `description`: Detailed text description
- `images`: Optional list of image data (multimodal)
- `audio_transcript`: Optional transcribed audio
- `reporter_contact`: Emergency contact info

---

### ğŸ“¤ Output Models

**1. TriageAssessment** (from Triage Agent)
- Disaster type classification
- Severity level (1-5 scale)
- Affected population estimate
- Immediate threats identified
- Priority actions recommended
- Confidence score

**2. ResourceAllocation** (from Resource Agent)
- Medical teams required
- Ambulances needed
- Fire units deployed
- Rescue teams assigned
- Shelters to open
- Food/water supplies
- Estimated cost

**3. CoordinationPlan** (from Coordination Agent)
- Primary responders assigned
- Secondary responders on standby
- Evacuation routes established
- Command center location
- Response timeline with phases

**4. PublicAlert** (from Communication Agent)
- Alert level (Warning/Emergency/Critical)
- Attention-grabbing headline
- Clear, concise message
- Step-by-step safety instructions

---

### ğŸ”„ Data Flow

DisasterReport
â†“
Intelligence Analysis
â†“
TriageAssessment
â†“
ResourceAllocation
â†“
CoordinationPlan + TriageAssessment
â†“

Each stage produces structured, validated data for the next stage.

````python
# ==========================================
# DATA MODELS & ENUMS
# ==========================================

print("ğŸ—ï¸ Building Data Models\n")
print("="*60)

class DisasterType(Enum):
    """Enumeration of disaster types the system can handle"""
    EARTHQUAKE = "earthquake"
    FLOOD = "flood"
    WILDFIRE = "wildfire"
    HURRICANE = "hurricane"
    TSUNAMI = "tsunami"
    LANDSLIDE = "landslide"
    TORNADO = "tornado"
    VOLCANIC = "volcanic_eruption"
    UNKNOWN = "unknown"


class SeverityLevel(Enum):
    """Disaster severity on 1-5 scale"""
    MINOR = 1
    MODERATE = 2
    SEVERE = 3
    CRITICAL = 4
    CATASTROPHIC = 5


@dataclass
class DisasterReport:
    """Input: Incoming disaster report"""
    report_id: str
    timestamp: str
    location: str
    description: str
    images: Optional[List[str]] = None
    audio_transcript: Optional[str] = None
    reporter_contact: Optional[str] = None


@dataclass
class TriageAssessment:
    """Output: Triage Agent assessment"""
    disaster_type: DisasterType
    severity: SeverityLevel
    affected_population: int
    immediate_threats: List[str]
    priority_actions: List[str]
    confidence_score: float


@dataclass
class ResourceAllocation:
    """Output: Resource Agent calculation"""
    medical_teams: int
    ambulances: int
    fire_units: int
    rescue_teams: int
    shelters_needed: int
    food_packages: int
    water_supply_liters: int
    estimated_cost: float


@dataclass
class CoordinationPlan:
    """Output: Coordination Agent plan"""
    primary_responders: List[str]
    secondary_responders: List[str]
    evacuation_routes: List[str]
    command_center_location: str
    timeline: Dict[str, str]


@dataclass
class PublicAlert:
    """Output: Communication Agent alert"""
    alert_level: str
    headline: str
    message: str
    safety_instructions: List[str]
    evacuation_notice: bool


print("âœ… Data models defined successfully\n")
print("ğŸ“Š Models created:")
print("   â€¢ DisasterType (enum - 9 types)")
print("   â€¢ SeverityLevel (enum - 5 levels)")
print("   â€¢ DisasterReport (input model)")
print("   â€¢ TriageAssessment (output model)")
print("   â€¢ ResourceAllocation (output model)")
print("   â€¢ CoordinationPlan (output model)")
print("   â€¢ PublicAlert (output model)")
print("\n" + "="*60 + "\n")
````

# 5. Agent Implementation

- Sends prompts to Gemini

- Records every interaction

- Enables audit trails

- Clean, maintainable, production-ready code

````python
# ==========================================
# BASE AGENT CLASS
# ==========================================

print("ğŸ¤– Building Base Agent Architecture\n")
print("="*60)

class BaseAgent:
    """
    Base class for all SENTINELS agents
    
    Features:
    - Gemini API integration
    - Session management with history
    - Error handling
    - Consistent interface
    """
    
    def __init__(self, name: str, role: str, model_name: str = MODEL_NAME):
        """
        Initialize a new agent
        
        Args:
            name: Agent identifier (e.g., "TriageAgent")
            role: Agent's responsibility description
            model_name: Gemini model to use
        """
        self.name = name
        self.role = role
        self.model = genai.GenerativeModel(
            model_name=model_name,
            generation_config=GENERATION_CONFIG
        )
        self.chat = None
        self.history = []
        print(f"   âœ“ {name} initialized ({role})")
    
    def initialize_session(self, system_prompt: str):
        """
        Start a new conversation session
        
        Args:
            system_prompt: Instructions that define agent behavior
        """
        self.chat = self.model.start_chat(history=[])
        self.system_prompt = system_prompt
    
    def process(self, input_data: str) -> str:
        """
        Process input through Gemini and return response
        
        Args:
            input_data: Input text to process
            
        Returns:
            str: Gemini's response
        """
        # Combine system prompt with input
        prompt = f"{self.system_prompt}\n\nInput: {input_data}"
        
        # Send to Gemini
        response = self.chat.send_message(prompt)
        
        # Store in history (Session Management - ADK #4)
        self.history.append({
            "input": input_data,
            "output": response.text,
            "timestamp": time.time()
        })
        
        return response.text
    
    def get_history(self) -> List[Dict]:
        """
        Retrieve full conversation history
        
        Returns:
            List of all interactions with timestamps
        """
        return self.history


print("\nâœ… Base Agent class ready!")
print("ğŸ”§ Features implemented:")
print("   â€¢ Gemini API integration")
print("   â€¢ Session management (ADK #4)")
print("   â€¢ Conversation history tracking")
print("   â€¢ Consistent interface for all agents")
print("\n" + "="*60 + "\n")
````

````python
# ==========================================
# TRIAGE & RESOURCE AGENTS
# ==========================================

print("ğŸ—ï¸ Building Specialized Agents (Part 1)\n")
print("="*60)

class TriageAgent(BaseAgent):
    """
    ğŸ¥ TRIAGE AGENT
    Analyzes disasters and classifies severity/type
    """
    
    def __init__(self):
        super().__init__(
            name="TriageAgent",
            role="Disaster Analysis & Classification"
        )
        self.initialize_session(self._get_system_prompt())
    
    def _get_system_prompt(self) -> str:
        return """You are a Disaster Triage Specialist AI Agent with emergency medicine expertise.

RESPONSIBILITIES:
1. Analyze disaster reports from multiple sources
2. Classify disaster type accurately (earthquake, flood, wildfire, etc.)
3. Assess severity on 1-5 scale (1=minor, 5=catastrophic)
4. Estimate affected population using available data
5. Identify immediate life-threatening hazards
6. Recommend priority actions for first responders

OUTPUT FORMAT: JSON only, no other text
{
  "disaster_type": "earthquake|flood|wildfire|hurricane|tsunami|landslide|tornado|volcanic_eruption",
  "severity": 1-5,
  "affected_population": number,
  "immediate_threats": ["threat1", "threat2"],
  "priority_actions": ["action1", "action2"],
  "confidence_score": 0.0-1.0
}

GUIDELINES:
- Be conservative with severity (don't overestimate)
- Base population on location data and disaster scope
- List only actionable, specific threats
- Prioritize life safety over property
- Return valid JSON only"""
    
    def analyze_disaster(self, report: DisasterReport) -> TriageAssessment:
        """Analyze disaster report and return structured assessment"""
        input_text = f"""
DISASTER REPORT #{report.report_id}
â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”
Location: {report.location}
Timestamp: {report.timestamp}
Description: {report.description}
Audio Transcript: {report.audio_transcript or 'None provided'}
Images: {'Available' if report.images else 'None'}
â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”

Analyze this disaster and provide JSON assessment.
"""
        response = self.process(input_text)
        
        # Try to parse JSON response
        try:
            # Clean response (remove markdown code blocks if present)
            clean_response = response.strip()
            if clean_response.startswith("```json"):
                clean_response = clean_response[7:]
            if clean_response.startswith("```"):
                clean_response = clean_response[3:]
            if clean_response.endswith("```"):
                clean_response = clean_response[:-3]
            clean_response = clean_response.strip()
            
            data = json.loads(clean_response)
            
            return TriageAssessment(
                disaster_type=DisasterType(data.get('disaster_type', 'unknown')),
                severity=SeverityLevel(int(data.get('severity', 3))),
                affected_population=int(data.get('affected_population', 100)),
                immediate_threats=data.get('immediate_threats', []),
                priority_actions=data.get('priority_actions', []),
                confidence_score=float(data.get('confidence_score', 0.5))
            )
        except Exception as e:
            print(f"   âš ï¸  JSON parsing failed, using fallback: {str(e)}")
            return self._fallback_parse(response, report)
    
    def _fallback_parse(self, response: str, report: DisasterReport) -> TriageAssessment:
        """Intelligent fallback if JSON parsing fails"""
        # Extract disaster type from response or description
        disaster_type = DisasterType.UNKNOWN
        for dtype in DisasterType:
            if dtype.value in response.lower() or dtype.value in report.description.lower():
                disaster_type = dtype
                break
        
        return TriageAssessment(
            disaster_type=disaster_type,
            severity=SeverityLevel.SEVERE,
            affected_population=1000,
            immediate_threats=["Multiple hazards present"],
            priority_actions=["Deploy emergency response", "Assess situation"],
            confidence_score=0.6
        )


class ResourceAgent(BaseAgent):
    """
    ğŸ“¦ RESOURCE AGENT
    Calculates emergency resources needed
    """
    
    def __init__(self):
        super().__init__(
            name="ResourceAgent",
            role="Resource Allocation & Logistics"
        )
        self.initialize_session(self._get_system_prompt())
    
    def _get_system_prompt(self) -> str:
        return """You are a Resource Allocation Specialist with emergency management expertise.

RESPONSIBILITIES:
1. Calculate required emergency personnel and equipment
2. Estimate medical resources (teams, ambulances)
3. Determine shelter and supply requirements
4. Provide realistic cost estimates

OUTPUT FORMAT: JSON only
{
  "medical_teams": number,
  "ambulances": number,
  "fire_units": number,
  "rescue_teams": number,
  "shelters_needed": number,
  "food_packages": number,
  "water_supply_liters": number,
  "estimated_cost": number
}

CALCULATION GUIDELINES:
- Medical teams: 1 per 500 people (severity multiplier)
- Ambulances: 1 per 1000 people (severity multiplier)
- Fire units: 1 per 2000 people (severity multiplier)
- Rescue teams: 2 per 1000 people (severity multiplier)
- Shelters: 1 per 100 people
- Food: 3 meals per person per day Ã— 3 days
- Water: 5 liters per person per day Ã— 3 days
- Cost: $500 per person Ã— severity level

Return valid JSON only."""
    
    def calculate_resources(self, assessment: TriageAssessment) -> ResourceAllocation:
        """Calculate resources based on triage assessment"""
        input_text = f"""
DISASTER ASSESSMENT
â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”
Type: {assessment.disaster_type.value}
Severity: {assessment.severity.value}/5 ({'âš ï¸' * assessment.severity.value})
Affected Population: {assessment.affected_population:,}
Immediate Threats: {', '.join(assessment.immediate_threats)}
â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”

Calculate required emergency resources in JSON format.
"""
        response = self.process(input_text)
        
        try:
            # Clean JSON
            clean_response = response.strip()
            if "```" in clean_response:
                clean_response = clean_response.split("```")[1]
                if clean_response.startswith("json"):
                    clean_response = clean_response[4:]
            clean_response = clean_response.strip()
            
            data = json.loads(clean_response)
            
            return ResourceAllocation(
                medical_teams=int(data.get('medical_teams', 1)),
                ambulances=int(data.get('ambulances', 1)),
                fire_units=int(data.get('fire_units', 1)),
                rescue_teams=int(data.get('rescue_teams', 2)),
                shelters_needed=int(data.get('shelters_needed', 1)),
                food_packages=int(data.get('food_packages', 100)),
                water_supply_liters=int(data.get('water_supply_liters', 500)),
                estimated_cost=float(data.get('estimated_cost', 10000))
            )
        except:
            return self._calculate_fallback(assessment)
    
    def _calculate_fallback(self, assessment: TriageAssessment) -> ResourceAllocation:
        """Mathematical fallback calculation"""
        pop = max(assessment.affected_population, 100)
        severity = assessment.severity.value
        
        return ResourceAllocation(
            medical_teams=max(1, (pop // 500) * severity),
            ambulances=max(1, (pop // 1000) * severity),
            fire_units=max(1, (pop // 2000) * severity),
            rescue_teams=max(2, (pop // 1000) * severity),
            shelters_needed=max(1, pop // 100),
            food_packages=pop * 3 * 3,  # 3 meals Ã— 3 days
            water_supply_liters=pop * 5 * 3,  # 5L Ã— 3 days
            estimated_cost=pop * 500 * severity
        )


print("\nâœ… Part 1 Agents ready!")
print("   ğŸ¥ TriageAgent - Disaster classification")
print("   ğŸ“¦ ResourceAgent - Resource calculation")
print("\n" + "="*60 + "\n")
````

````python
# ==========================================
# ORCHESTRATOR - THE BRAIN
# ==========================================

print("ğŸ­ Building Master Orchestrator\n")
print("="*60)

class DisasterSentinelOrchestrator:
    """
    Main orchestrator coordinating all 5 agents
    Implements ADK Concept #2: Agent Orchestration
    """
    
    def __init__(self):
        print("ğŸš€ Initializing SENTINELS Multi-Agent System...\n")
        
        # Initialize all 5 specialized agents
        self.triage_agent = TriageAgent()
        self.resource_agent = ResourceAgent()
        self.coordination_agent = CoordinationAgent()
        self.communication_agent = CommunicationAgent()
        self.intelligence_agent = IntelligenceAgent()
        
        print("\n" + "="*60)
        print("âœ… All 5 agents initialized successfully!")
        print("="*60)
        
        # Session data storage
        self.session_data: List[Dict[str, Any]] = []
    
    def process_disaster_report(self, report: DisasterReport) -> Dict[str, Any]:
        """
        MAIN PROCESSING PIPELINE
        Orchestrates all agents in sequence
        
        Args:
            report: DisasterReport object
            
        Returns:
            Complete response dictionary
        """
        print(f"\n{'='*60}")
        print(f"ğŸš¨ PROCESSING DISASTER REPORT #{report.report_id}")
        print(f"{'='*60}\n")
        print(f"ğŸ“ Location: {report.location}")
        print(f"â° Time: {report.timestamp}\n")
        
        # Initialize results dictionary
        results: Dict[str, Any] = {
            "report_id": report.report_id,
            "timestamp": report.timestamp,
            "location": report.location,
            "processing_time": {}
        }
        
        start_time = time.time()
        
        # ==========================================
        # STEP 1: INTELLIGENCE ANALYSIS
        # ==========================================
        print("ğŸ“Š Step 1: Intelligence Analysis...")
        step_start = time.time()
        
        try:
            if report.audio_transcript:
                audio_analysis = self.intelligence_agent.analyze_audio_transcript(
                    report.audio_transcript
                )
                results["audio_analysis"] = audio_analysis
                print(f"   âœ“ Audio analyzed - Urgency: {audio_analysis.get('urgency', 'N/A')}")
            
            if report.images:
                image_analysis = self.intelligence_agent.analyze_image(
                    report.images[0], report.description
                )
                results["image_analysis"] = image_analysis
                print(f"   âœ“ Image analyzed")
        except Exception as e:
            print(f"   âš ï¸  Intelligence analysis error: {str(e)}")
            results["intelligence_error"] = str(e)
        
        results["processing_time"]["intelligence"] = round(time.time() - step_start, 2)
        
        # ==========================================
        # STEP 2: TRIAGE ASSESSMENT
        # ==========================================
        print("\nğŸ¥ Step 2: Triage Assessment...")
        step_start = time.time()
        
        try:
            assessment: TriageAssessment = self.triage_agent.analyze_disaster(report)
            results["triage"] = assessment  # store dataclass, not dict
            
            print(f"   âœ“ Type: {assessment.disaster_type.value}")
            print(f"   âœ“ Severity: {assessment.severity.value}/5 {'âš ï¸' * assessment.severity.value}")
            print(f"   âœ“ Affected: {assessment.affected_population:,} people")
            print(f"   âœ“ Confidence: {assessment.confidence_score:.1f}%")
        except Exception as e:
            print(f"   âŒ Triage error: {str(e)}")
            raise
        
        results["processing_time"]["triage"] = round(time.time() - step_start, 2)
        
        # ==========================================
        # STEP 3: RESOURCE ALLOCATION
        # ==========================================
        print("\nğŸ“¦ Step 3: Resource Allocation...")
        step_start = time.time()
        
        try:
            resources: ResourceAllocation = self.resource_agent.calculate_resources(assessment)
            results["resources"] = resources  # store dataclass
            
            print(f"   âœ“ Medical Teams: {resources.medical_teams}")
            print(f"   âœ“ Ambulances: {resources.ambulances}")
            print(f"   âœ“ Rescue Teams: {resources.rescue_teams}")
            print(f"   âœ“ Shelters: {resources.shelters_needed}")
            print(f"   âœ“ Estimated Cost: ${resources.estimated_cost:,.0f}")
        except Exception as e:
            print(f"   âŒ Resource calculation error: {str(e)}")
            raise
        
        results["processing_time"]["resources"] = round(time.time() - step_start, 2)
        
        # ==========================================
        # STEP 4: COORDINATION PLANNING
        # ==========================================
        print("\nğŸ¯ Step 4: Coordination Planning...")
        step_start = time.time()
        
        try:
            plan: CoordinationPlan = self.coordination_agent.create_plan(assessment, resources)
            results["coordination"] = plan  # store dataclass
            
            print(f"   âœ“ Primary: {', '.join(plan.primary_responders[:3])}")
            print(f"   âœ“ Command Center: {plan.command_center_location}")
            print(f"   âœ“ Evacuation Routes: {len(plan.evacuation_routes)}")
        except Exception as e:
            print(f"   âŒ Coordination error: {str(e)}")
            raise
        
        results["processing_time"]["coordination"] = round(time.time() - step_start, 2)
        
        # ==========================================
        # STEP 5: PUBLIC COMMUNICATION
        # ==========================================
        print("\nğŸ“¢ Step 5: Public Alert Generation...")
        step_start = time.time()
        
        try:
            alert: PublicAlert = self.communication_agent.generate_alert(assessment, plan)
            results["alert"] = alert  # store dataclass
            
            print(f"   âœ“ Alert Level: {alert.alert_level}")
            print(f"   âœ“ Headline: {alert.headline}")
            print(f"   âœ“ Evacuation: {'YES' if alert.evacuation_notice else 'NO'}")
        except Exception as e:
            print(f"   âŒ Communication error: {str(e)}")
            raise
        
        results["processing_time"]["communication"] = round(time.time() - step_start, 2)
        
        # ==========================================
        # FINALIZE
        # ==========================================
        total_time = round(time.time() - start_time, 2)
        results["processing_time"]["total"] = total_time
        
        # Store in session
        self.session_data.append(results)
        
        print(f"\n{'='*60}")
        print(f"âœ… DISASTER RESPONSE COMPLETE")
        print(f"â±ï¸  Total Processing Time: {total_time}s")
        print(f"{'='*60}\n")
        
        return results
    
    def display_summary(self, results: Dict[str, Any]):
        """Display formatted summary of disaster response"""
        print("\n" + "â•”" + "="*58 + "â•—")
        print("â•‘" + " "*18 + "RESPONSE SUMMARY" + " "*24 + "â•‘")
        print("â•š" + "="*58 + "â•\n")
        
        # ---------------- TRIAGE SECTION ----------------
        triage: TriageAssessment = results["triage"]
        
        print("ğŸ¥ TRIAGE ASSESSMENT")
        print("â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€")
        print(f"   Type: {triage.disaster_type}")
        
        # Convert Enum to number safely
        severity_value = triage.severity.value
        print(f"   Severity: {severity_value}/5 {'âš ï¸' * severity_value}")
        print(f"   Affected Population: {triage.affected_population:,} people")
        print(f"   Immediate Threats: {', '.join(triage.immediate_threats[:3])}")
        print(f"   Priority Actions: {', '.join(triage.priority_actions[:3])}")
        print(f"   Confidence: {triage.confidence_score * 100:.1f}%")
        
        # ---------------- RESOURCES SECTION ----------------
        resources: ResourceAllocation = results["resources"]
        print(f"\nğŸ“¦ RESOURCES DEPLOYED")
        print("â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€")
        print(f"   ğŸ‘¨â€âš•ï¸ Medical Teams: {resources.medical_teams}")
        print(f"   ğŸš‘ Ambulances: {resources.ambulances}")
        print(f"   ğŸš’ Fire Units: {resources.fire_units}")
        print(f"   ğŸ†˜ Rescue Teams: {resources.rescue_teams}")
        print(f"   ğŸ  Shelters: {resources.shelters_needed}")
        print(f"   ğŸ’§ Water (L): {resources.water_supply_liters:,}")
        print(f"   ğŸ± Food Packages: {resources.food_packages:,}")
        print(f"   ğŸ’° Est. Cost: ${resources.estimated_cost:,.0f}")
        
        # ---------------- COORDINATION SECTION ----------------
        coordination: CoordinationPlan = results["coordination"]
        print(f"\nğŸ¯ COORDINATION PLAN")
        print("â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€")
        print(f"   Primary Responders: {', '.join(coordination.primary_responders)}")
        print(f"   Secondary Responders: {', '.join(coordination.secondary_responders)}")
        print(f"   Command Center: {coordination.command_center_location}")
        print(f"   Evacuation Routes: {len(coordination.evacuation_routes)}")
        
        # ---------------- ALERT SECTION ----------------
        alert: PublicAlert = results["alert"]
        print(f"\nğŸ“¢ PUBLIC ALERT")
        print("â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€")
        print(f"   Level: {alert.alert_level}")
        print(f"   Headline: {alert.headline}")
        print(f"   Evacuation Required: {'YES âš ï¸' if alert.evacuation_notice else 'NO'}")
        print(f"   Message: {alert.message}")
        
        # ---------------- PERFORMANCE SECTION ----------------
        times = results["processing_time"]
        print(f"\nâ±ï¸  PERFORMANCE")
        print("â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€")
        print(f"   Total Time: {times['total']}s")
        print(f"   Intelligence: {times.get('intelligence', 0)}s")
        print(f"   Triage: {times.get('triage', 0)}s")
        print(f"   Resources: {times.get('resources', 0)}s")
        print(f"   Coordination: {times.get('coordination', 0)}s")
        print(f"   Communication: {times.get('communication', 0)}s")
        print(f"   Target: <60s {'âœ…' if times['total'] < 60 else 'âš ï¸'}")
        
        print("\n" + "="*60 + "\n")


print("âœ… Orchestrator ready!")
print("ğŸ­ Coordinates all 5 agents in unified pipeline")
print("ğŸ“Š Implements complete disaster response workflow")
print("\n" + "="*60 + "\n")
````

````python
# ==========================================
# SESSION MEMORY SYSTEM
# ==========================================

class SessionMemory:
    """
    Stores and retrieves historical disaster responses
    Enables learning across sessions
    """

    def __init__(self):
        self.memory = []

    def store(self, response_data):
        self.memory.append(response_data)

    def retrieve_recent(self, limit=3):
        return self.memory[-limit:]

    def total_sessions(self):
        return len(self.memory)

print("âœ… Session Memory System Activated")

# Attach memory to orchestrator
DisasterSentinelOrchestrator.session_memory = SessionMemory()
print("âœ… Memory linked to Orchestrator")
````

## ğŸ›  Custom Emergency Intelligence Tools

Disaster Sentinel now integrates heuristic-based tools to support agents:

1. Risk Scoring Tool
2. Priority Ranking Tool
3. Emergency Impact Estimator

These tools improve accuracy and decision confidence.


````python
# ==========================================
# CUSTOM TOOLS MODULE
# ==========================================

class EmergencyTools:

    @staticmethod
    def calculate_risk(severity, population):
        """
        Risk score based on severity and population exposure
        """
        return min(100, (severity * population) / 1000)

    @staticmethod
    def priority_index(severity, confidence):
        """
        Combines urgency and model confidence
        """
        return round((severity * confidence) * 10, 2)

print("âœ… Custom Emergency Tools Activated")

# Attach tools to orchestrator
DisasterSentinelOrchestrator.tools = EmergencyTools
print("âœ… Emergency Tools Linked")
````

````python
# ==============================================
# ENHANCE RESULTS WITH RAG KNOWLEDGE
# ==============================================

def enhance_with_rag(results):
    triage = results.get("triage")

    if isinstance(triage, dict):
        disaster_type = triage.get("disaster_type", "unknown")
    else:
        disaster_type = getattr(triage, "disaster_type", "unknown")

    dtype = str(disaster_type).lower().split(".")[-1]
    results["rag_guidance"] = retrieve_rag_guidance(dtype)

    print(f"ğŸ“š RAG guidance attached for: {dtype}")
    return results
````
